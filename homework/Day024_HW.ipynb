# Julia 基礎圖表繪製：圖表結合敘述統計

## 作業 024：繪製帶統計資訊的圖表

繪製 iris 資料的散佈圖，x 軸為 PetalLength，y 軸為 PetalWidth，並依不同花的種類分別算出 PetalLength 及 PetalWidth 的平均值，將平均值以鉛直線與水平線標在圖上。（請完成以下程式碼）

```julia
using StatsBase, RDatasets
using Gadfly
```

```julia
iris = dataset("datasets", "iris")
```

```julia
petalmean = by(iris,:Species,PetalLength = :PetalLength=>mean,PetalWidth = :PetalWidth=>mean)
```

```julia
p = plot(iris ,xgroup=:Species,x=:PetalLength,y=:PetalWidth,Geom.subplot_grid(
        layer(petalmean,xgroup=:Species,xintercept=:PetalLength,Geom.vline(style=:dot,color = "red")),
        layer(petalmean,xgroup=:Species,yintercept=:PetalWidth,Geom.hline(style=:dot,color = "green")),
        layer(Geom.point)))
```
